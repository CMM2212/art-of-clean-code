
# WHat is flow?
Flow is a state of pure concentration. 

State of complete immersion in the task at hand

six components:

1. attention: deep sense of focus
2. action: bias towards forward action
3. self: less aware of self, more aware of action
4. control: increased control
5. time: lose ability to experience time passing
6. reward: the labor of the activity is your desire

# How to achieve flow

1. goal must be clear
2. feedback must be immediate
3. balance between opportunity and capacity

## clear goals

must have a clear project goal. every line of code must bring you closer to that goal

## feedback mechanism

seek more feedback of oters, publish code, share MVP early and often. 

## Balance opportunity and capacity

constantly seek harder challenges without reaching anxiety levels.

# Flow tips for coders

1. know what to do
2. know how to do it.
3. know how well you're doing
4. know where to go
5. seek challenges
6. work on your skills to overcome the challenges
7. free yourself from distractions

## other tactics

1. always have a practical code project in the works. 70/30 working on projects vs reading books

2. work on fun projects that fufill your purpose. 

3. perform from your strengths

4. block otu your coding time in large chunks

5. eliminate distractions during yoru flow time

6. do obvious things you know you need to do. eat sleep, exercise

7. consume high quality information: read books instead of shallow blogs. or read research papers
