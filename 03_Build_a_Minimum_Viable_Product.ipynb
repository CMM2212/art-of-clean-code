MVP: A version of your product stripped of all except the most necessary features, in order to test and validate your hyptheses quickly.

# Stealth mode issues

Stealth mode being when you do things alone


## Loss of motivation

Doubts grow. Similar tools are found to exist. Belief that it cannot be done arises. Loss of motivation can kill your product

On the other hand if you release an early version of the tool, feedback may inspireyou to improve or overcome problems.

## Distraction

MVP can combat distractions by creating more immediate feedback that helps refocus your attention



## Running over time

Fewer features = fewer things go wrong.

The more predictable your project, the more you will believe in its success.

## Lack of response

MVP-based approach helps develop the project to directly address the most pressing customer needs, increasing chances of your customer engagement and therefore response to early product versions



## Wrong Assumptions

Without feedback, you may have the wrong assumptions from users



## Unnecessary Complexity

By releasing all features at once, you don't know if the individual features may have been more popular.

# Building an MVP

Form a hypothesis -> create product that validates hypothesis. Remove all features that don't validate this hypothesis.

Once your first MVP is analyzed, you can build a second MVP that adds the next feature.

This is called rapid prototyping: release early and often

# Four Pillars of Building an MVP


## Functionality

Product provides a clear function to the user. Doesn't have to be efficient.



## Design

Focused design that supports your target niche

## Reliability

Write tests.

Your feedback may otherwise become just about complaints of reliability rather than the feature.


## Usability

It should be obvious how to use. Minimal amount of buttons/fields

# Myth around stealing

People sometimes use stealth mode to protect their ideas. 

Ideas are cheap and not special. Execution is everything.

Keeping ideas secret = bad
