# Material Design

Philosophy developed by gogle which describes a way to organize and design screen elements according to what users already understand: physical world elements such as paper, cards, pens, and shadow.

# How to achieve minimalistic design


## use whitespace

one of the key pillars. Rpleacing design elements with whitespace will improve clarity.

scrolling is cheap, spread out your text.

## remove design elements

go over each design element and throw it out of possible.

design elements are any visible element of the UI (menu, calsl to action, lists, buttons, images, boxes, shadows, videos, etc).

Be bold, remove important design elements and leave only very important elements

## remove features

don't allow feature creep. keep your features limited and stick to what is most important


## reduce variations of fonts/colors

two font types, one or two colors, one or two font sizes

## Be consistent

use consistent templates and CSS
