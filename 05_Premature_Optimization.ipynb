# Tips for performance Tuning


## Measure first, improve second

start with a naive approach and measure its performance, then compare it to a different approach

## Pareto is king

20% of the code is likely causing 80% of the performance issues


## Algorithmic Optimization wins

Try imrpvoed algorithms or better data structures

## All hail the cache

caching allows you to store previous computations

# Less is more

Instead of optimizing, reduce the the complexity and get rid of unnecessary features and computations.

* Remove your current bottleneck altogether by just skipping the feature
* Replace problem with a simpler version
* Omit one important/expensive feature, so you can pursue more important ones
