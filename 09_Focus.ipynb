# The Weapon against Complexity

complexity leads to chaos

use focus to move from high entropy to low entropy. Plan carefully and focus your effort.

# Unifying the Principles


## 80/20 principle

focus on the vital few important things


## Build a minimal viable product

focus on one hypothesis at a time. before you begin have a clear hypothesis on what the program should solve. remove all unecessary features


release your mvp early and often


## write clean and simple code

code is read more than written


## premature optimization is the root of all evil

forget about small inefficienies


## flow

flow is a state of focus

## do one thing well (unix)

programs should do one thing well. to do something bigger, you should combine these programs.

## less is more in design

remove all unecessary features and design. keep things simple

# conclusion

ask yourself:

* On which software project do I want to focus my efforts?

* Which features do I want to focus on to create my MVP?

* What is the minimal number of design elements I can implement to test the viability of my product?

* Who will use my product and why?

* What can I remove from my code?

* Do my functions do one thing only?

* How can I achieve the same result in less time
